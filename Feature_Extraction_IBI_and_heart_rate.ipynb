```python
import pandas as pd
import numpy as np
import os
from google.colab import drive

video_names = ['A1', 'A2', 'A3', 'A4', 'A', 'B', 'C', 'F', 'G', 'H', 'J', 'K', 'M', 'N', 'O', 'P', 'Q', 'U', 'V', 'W']
data_frames = []

# Mounting Google Drive
drive.mount('/content/drive')

# Redefining the file path for Google Drive
input_file_path = '/content/drive/My Drive/OutputVF/interpolated_vF_16.csv'  # Path to your input file
output_file_path = '/content/drive/My Drive/ResultsVF/results_vF_16.csv'  # Path to save your output file

# Load the dataset
df = pd.read_csv(input_file_path, low_memory=False)

# new_header = df.iloc[29]
# df = df.iloc[30:]
# df.columns = new_header
# df.dropna(axis=1, how="all", inplace=True)
df['Heart Rate PPG ALG'].replace(-1, np.nan, inplace=True)
df['IBI PPG ALG'].replace(-1, np.nan, inplace=True)
df['Heart Rate PPG ALG'] = pd.to_numeric(df['Heart Rate PPG ALG'], errors='coerce')
df['IBI PPG ALG'] = pd.to_numeric(df['IBI PPG ALG'], errors='coerce')
#print(df[['Heart Rate PPG ALG', 'IBI PPG ALG']])

video_differences = []
fixation_data = df[df['SourceStimuliName'] == 'Fixation point']
fixation_data_ibi = fixation_data[(fixation_data['IBI PPG ALG'].notnull())]
# print(fixation_data_ibi)



if not fixation_data.empty:
    mean_ibi_fixation = fixation_data_ibi['IBI PPG ALG'].mean()
    print(mean_ibi_fixation)
    for video in video_names:
        video_data = df[df['SourceStimuliName'] == video]
        #video_data = video_data[(video_data['Heart Rate PPG ALG'].notnull()) & (video_data['IBI PPG ALG'].notnull())]


        if not video_data.empty:
            ibi_value = video_data['IBI PPG ALG']
            each_video_normalize_ibi = (ibi_value - mean_ibi_fixation) / mean_ibi_fixation
            # print(each_video_normalize_ibi)

            # if len(ibi_value) < len(mean_ibi_fixation):
            #     each_video_normalize_ibi = np.random.permutation(mean_ibi_fixation)
            #     each_video_normalize_ibi = each_video_normalize_ibi[: len(mean_ibi_fixation)]
            # print(each_video_normalize_ibi)

            difference = np.mean(each_video_normalize_ibi)
            min_difference = np.min(each_video_normalize_ibi)
            max_difference = np.max(each_video_normalize_ibi)
            skewness = pd.Series(each_video_normalize_ibi).skew()
            kurtosis = pd.Series(each_video_normalize_ibi).kurtosis()
            std_dev = np.std(each_video_normalize_ibi)

            column_name_mapping = {
                'A1': 'A1_LP',
                'A2': 'A2_LP',
                'A3': 'A3_LP',
                'A4': 'A4_LP',
                'A': 'A_HN',
                'B': 'B_HN',
                'C': 'C_LN',
                'F': 'F_HN',
                'G': 'G_HP',
                'H': 'H_HP',
                'J': 'J_Ne',
                'K': 'K_Ne',
                'M': 'M_LN',
                'N': 'N_LN',
                'O': 'O_LN',
                'P': 'P_HP',
                'Q': 'Q_HP',
                'U': 'U_Ne',
                'V': 'V_Ne',
                'W': 'W_HN'
            }

            video_differences.append({
                #'participant name': participant,
                'video name': video,
                'ibi mean normalize': difference,
                'ibi min normalize': min_difference,
                'ibi max normalize': max_difference,
                'ibi skew normalize': skewness,
                'ibi kurtosis normalize': kurtosis,
                'ibi std normalize': std_dev
            })

    # Create DataFrame for each participant and append to data_frames
    result_df = pd.DataFrame(video_differences)
    result_df['video name'] = result_df['video name'].replace(column_name_mapping)
    data_frames.append(result_df)

# Concatenate all DataFrames into one outside the loop
merged_data = pd.concat(data_frames, ignore_index=True)

# Saving the processed DataFrame to the Output folder in Google Drive
merged_data.to_csv(output_file_path, index=False)
```
